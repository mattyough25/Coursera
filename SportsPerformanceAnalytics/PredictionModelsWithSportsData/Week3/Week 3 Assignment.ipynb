```python
import pandas as pd
import numpy as np
```

```python
BLmod = pd.read_excel("D:\\GitHub\\Coursera\\SportsPerformanceAnalytics\\PredictionModelsWithSportsData\\data\\Bundesliga prediction model (Assignment).xlsx")
BLmod
```

```python
BLmod['B365HPr']= 1/(BLmod['B365H'])/(1/(BLmod['B365H'])+ 1/(BLmod['B365D'])+ 1/(BLmod['B365A']))
BLmod['B365DPr']= 1/(BLmod['B365D'])/(1/(BLmod['B365H'])+ 1/(BLmod['B365D'])+ 1/(BLmod['B365A']))
BLmod['B365APr']= 1/(BLmod['B365A'])/(1/(BLmod['B365H'])+ 1/(BLmod['B365D'])+ 1/(BLmod['B365A']))
BLmod
```

```python
BLmod['B365_result'] = np.where((BLmod['B365HPr'] > BLmod['B365APr']) & (BLmod['B365HPr'] > BLmod['B365DPr']),'H', (np.where((BLmod['B365APr'] > BLmod['B365HPr']) & (BLmod['B365APr'] > BLmod['B365DPr']),'A','D')))
BLmod
```

```python
BLmod['win_value'] = np.where(BLmod['FTHG'] > BLmod['FTAG'] ,2, (np.where(BLmod['FTHG'] < BLmod['FTAG'],0,1)))
BLmod
```

```python
BLmod['Home_Win'] = np.where(BLmod['FTR'] == 'H',1,0)
BLmod['Away_Win'] = np.where(BLmod['FTR'] == 'A',1,0)
BLmod
```

```python
pd.crosstab(BLmod['FTR'], BLmod['B365_result'],dropna= True)
```

```python
BLmod['lhTMratio'] = np.log(BLmod['Tmhome']/BLmod['Tmaway'])
```

```python
BL_first = BLmod[:224]
BL_first
```

```python
from bevel.linear_ordinal_regression import OrderedLogit
ol = OrderedLogit()
```

```python
ol.fit(BL_first['lhTMratio'], BL_first['win_value'])
ol.print_summary()
```

```python
print(f'beta = {ol.coef_[0]:.4f}')
print(f'interceptAD = {ol.coef_[1]:.4f}')
print(f'interceptDH = {ol.coef_[2]:.4f}')
```

```python
BLmod['predA'] = 1/(1+np.exp(-(ol.coef_[1]-ol.coef_[0]*BLmod['lhTMratio'])))
BLmod['predD'] = 1/(1+np.exp(-(ol.coef_[2]-ol.coef_[0]*BLmod['lhTMratio']))) - BLmod['predA']
BLmod['predH'] = 1 - BLmod['predA'] - BLmod['predD']

pd.set_option('display.max_columns', 50)
BLmod
```

```python
BLmod['Maxprob'] =BLmod[['predA','predD','predH']].max(axis=1)
BLmod['logitpred']=np.where(BLmod['Maxprob']==BLmod['predA'],'A',\
                               np.where(BLmod['Maxprob']==BLmod['predD'],'D','H'))
BLmod['logittrue']= np.where(BLmod['logitpred'] == BLmod['FTR'],1,0)
BLmod
```

```python
BLmod['Draw'] = np.where(BLmod['FTR'] == 'D',1,0)
BLmod
```
