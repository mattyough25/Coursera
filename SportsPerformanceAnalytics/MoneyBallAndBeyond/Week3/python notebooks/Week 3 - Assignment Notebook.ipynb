# II. Running Regressions for Each Season

```python
# Write a function to run the Moneyball regression annually for free agents only
def MBExpandFA(Season):
    MB_Seas = Master_95_15[(Master_95_15.SalYear == Season) & (Master_95_15.Free == 1)]
    global lmrate
    lmrate = smf.ols(formula = 'lnSal ~ BA + ISO_POWER_x + EYE + PA + Exp + Exp2 + C(POS)', data=MB_Seas).fit()
    return;
```

```python
#Your Code Here
index = 0
lmrate_Results = [0]
for index in range(1,21):
    lmrate_Results.append(index)
    index = index + 1
display(lmrate_Results) 
```

```python
Season = 1995
i = 0
while Season <= 2015:
    MBExpandFA(Season)
    lmrate_Results[i] = lmrate
    i = i + 1
    Season = Season + 1
```

# III. Running the Pooled Regression

```python
#Your Code Here
Master_Free = Master_95_15[Master_95_15.Free == 1].copy()
Master_Free['PreMB'] = np.where(Master_Free['SalYear']<=2004,1,0)
Master_Free.describe()
```

```python
Pooled_lm = smf.ols(formula = 'lnSal ~ BA + ISO_POWER_x + EYE + PA + Exp + Exp2 + C(POS) + PreMB*(BA + ISO_POWER_x + EYE + PA + Exp + Exp2 + C(POS))',\
                    data=Master_Free).fit()
Pooled_lm.summary()
```
