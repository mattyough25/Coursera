```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
import statsmodels.formula.api as smf
```

```python
NBAmod = pd.read_excel("D:\\GitHub\\Coursera\\SportsPerformanceAnalytics\\PredictionModelsWithSportsData\\data\\NBA prediction model (Assignment).xlsx")
NBAmod
```

```python
# Win probabilities
NBAmod['prob_hwin']= 1/(NBAmod['hwinodds'])/(1/(NBAmod['hwinodds'])+ 1/(NBAmod['hloseodds']))
NBAmod['prob_awin']= 1/(NBAmod['hloseodds'])/(1/(NBAmod['hloseodds'])+ 1/(NBAmod['hwinodds']))
```

```python
# Define Home Loss Variable
NBAmod['hlose'] = np.where(NBAmod['hwin'] != 1, 1,0)
```

```python
# Winning Team Variable
NBAmod['winning_team'] = np.where(NBAmod['hwin'] == 1, 'H','A')
NBAmod
```

```python
# Define Team with Highest Win Probability
NBAmod['prob_winning_team'] = np.where(NBAmod['prob_hwin'] > 0.5, 'H','A')
```

```python
# Log of Salary Variable
NBAmod['lg_home_sal']=np.log(NBAmod['hteamsal'])
NBAmod['lg_away_sal']=np.log(NBAmod['opposal'])

# taking salary ratio between two teams
NBAmod['lg_sal_ratio']= NBAmod['lg_home_sal'] - NBAmod['lg_away_sal']
```

```python
# Get 2018 Data
NBA18 = NBAmod[NBAmod.year == 2018]
NBA18
```

```python
# Logistic Regression
Win_lgsal = 'hwin~lg_sal_ratio'
logit_model = smf.glm(formula = Win_lgsal, data = NBA18, family=sm.families.Binomial())
result = logit_model.fit()
print(result.summary())
```

```python
# Define Predicted Probabilities and Predicted Results
NBAmod['predHW'] = 1/(1+np.exp((.4452+1.1216*NBAmod['lg_sal_ratio'])))
NBAmod['predAW'] = 1- NBAmod['predHW'] 
NBAmod
```

```python
#Define Predicted Result
NBAmod['pred_winning_team'] = np.where(NBAmod['predHW'] > 0.5, 'H','A')
```

```python
# Get 2019 Data
NBA19 = NBAmod[NBAmod.year == 2019]
NBA19
```

```python
# Determine if Bookmaker was right
NBA19['BM_Right'] = np.where(NBA19['winning_team'] == NBA19['prob_winning_team'], 1, 0)
NBA19
```

```python
# Determine if Model was right
NBA19['Logit_Right'] = np.where(NBA19['winning_team'] == NBA19['pred_winning_team'], 1, 0)
NBA19
```

```python
# Logit Brier Score
NBA19['Houtcome'] = np.where(NBA19['winning_team'] == 'H', 1,0)
NBA19['Aoutcome'] = np.where(NBA19['winning_team'] == 'A', 1,0)

BrierLogit = ((NBA19['predHW'] - NBA19['Houtcome'])**2 + (NBA19['predAW'] - NBA19['Aoutcome'])**2).sum()/688

BrierLogit
```

```python
# Bookmaker Brier Score
BrierBM = ((NBA19['prob_hwin'] - NBA19['Houtcome'])**2 + (NBA19['prob_awin'] - NBA19['Aoutcome'])**2).sum()/688

BrierBM
```
