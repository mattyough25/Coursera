# I. Data Coding and Merging

```python
# Load packages
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Load game logs data
Teams = pd.read_excel("D:\\GitHub\\Coursera\\SportsPerformanceAnalytics\\MoneyBallAndBeyond\\data\\Game logs 1999-2003.xlsx")
```

```python
#Your Code Here
Teams['hwin'] = np.where(Teams['home_score'] > Teams['visitor_score'],1,0)
Teams['awin'] = np.where(Teams['home_score'] < Teams['visitor_score'],1,0)
```

```python
Teams['year'] = Teams['date'].astype(str).str[0:4]
display(Teams)
```

```python
Teamshome = Teams.groupby(['home','year'])['visitor_ab', 'visitor_h', 'visitor_2b', 'visitor_3b', 'visitor_hr', 'visitor_sf',
                                           'visitor_hbp','visitor_bb','home_ab', 'home_h', 'home_2b', 'home_3b',
                                           'home_hr', 'home_sf', 'home_hbp', 'home_bb', 'hwin','awin'].sum()
Teamshome = Teamshome.reset_index()
Teamshome = Teamshome.rename(columns = {"home": "Team"}) 
Teamshome
```

```python
Teamsaway = Teams.groupby(['year','visitor'])['visitor_ab', 'visitor_h', 'visitor_2b', 'visitor_3b', 
                                           'visitor_hr', 'visitor_sf', 'visitor_hbp', 'visitor_bb', 
                                           'home_ab', 'home_h', 'home_2b', 'home_3b', 'home_hr', 
                                           'home_sf', 'home_hbp', 'home_bb','hwin','awin'].sum()

Teamsaway = Teamsaway.reset_index()
Teamsaway = Teamsaway.rename(columns = {"visitor": "Team"}) 
Teamsaway
```

```python
Teams2 = pd.merge(Teamshome, Teamsaway, on =['Team','year'])
pd.set_option('display.max_columns', 50)
Teams2['wins'] = Teams2['hwin_x']+Teams2['awin_y']
display(Teams2)
```

```python
Teams2.groupby(['year']).describe()
```

```python
Hits = Teams2.groupby('year')['home_h_x','visitor_h_y'].sum()
Hits['Total'] = Hits['home_h_x'] + Hits['visitor_h_y']
display(Hits)
```

```python
AB = Teams2.groupby('year')['visitor_ab_x','home_ab_y'].sum()
AB['Total'] = AB['visitor_ab_x'] + AB['home_ab_y']
AB['avg_ab'] = AB['Total']/162
display(AB)
```

# II. Calculating Performance Metrics

```python
#Your Code Here
Teams2['BAFOR'] = ((Teams2['home_h_x']+Teams2['visitor_h_y'])/
                    (Teams2['home_ab_x']+Teams2['visitor_ab_y']))
```

```python
Teams2['BAAGN'] = ((Teams2['home_h_y']+Teams2['visitor_h_x'])/
                    (Teams2['home_ab_y']+Teams2['visitor_ab_x']))
```

```python
Teams3 = Teams2[['year', 'Team', 'wins','BAFOR','BAAGN']]
```

```python
TeamsGh = Teams.groupby(['year','home'])['hwin'].count()
TeamsGh = TeamsGh.reset_index()
TeamsGh = TeamsGh.rename(columns = {"home": 'Team',"hwin": 'hgames'}) 

TeamsGa = Teams.groupby(['year','visitor'])['awin'].count()
TeamsGa = TeamsGa.reset_index()
TeamsGa = TeamsGa.rename(columns = {"visitor": 'Team','awin':'agames'}) 

TeamsG = pd.merge(TeamsGh, TeamsGa, on =['Team','year'])
TeamsG['Games'] = TeamsG['hgames']+TeamsG['agames']
TeamsG = TeamsG[['year','Team','Games','hgames','agames']]
TeamsG
```

```python
TeamsGh2 = Teamshome[['year','Team','hwin']]

TeamsGa2 = Teamsaway[['year','Team','awin']]

TeamsG2 = pd.merge(TeamsGh2, TeamsGa2, on =['Team','year'])
#TeamsG = TeamsG[['year','Team','Games','hgames','agames']]
TeamsG2
```

```python
Teams3 = pd.merge(Teams3, TeamsG, on =['Team','year'])
```

```python
Teams3
```

```python
Teams4 = pd.merge(Teams3, TeamsG2, on =['Team','year'])
```

```python
Teams4
```

```python
Teams4['wpc'] = Teams4['wins']/Teams4['Games']
Teams4
```

```python
Teams4['h_wpc'] = Teams4['hwin']/Teams4['hgames']
Teams4['a_wpc'] = Teams4['awin']/Teams4['agames']
Teams4
```

```python
Teams4['WPCT_diff'] = abs(Teams4['h_wpc'] - Teams4['a_wpc'])
Teams4
```

```python
ix = Teams4.index[Teams4['WPCT_diff'] == max(Teams4['WPCT_diff'])].to_list()

Max_Team = Teams4.iloc[ix[0]]['Team']
Max_Year = Teams4.iloc[ix[0]]['year']
A_PCT    = Teams4.iloc[ix[0]]['a_wpc']
A_PCT
```

```python
Teams4['AVG_Diff'] = Teams4['BAFOR'] - Teams4['BAAGN']
Teams4
```

```python
print(min(Teams4['AVG_Diff']))
```

```python
Teams4['NOT_162'] = np.where(Teams4['Games'] != 162,1,0)
Teams4
```

# III. Running Regressions

```python
#Your Code Here
import statsmodels.formula.api as smf

WinBA_lm = smf.ols(formula = 'wpc ~ BAFOR + BAAGN', data=Teams4).fit()
WinBA_lm.summary()
```

```python
WinBA2_lm = smf.ols(formula = 'wpc ~ I(BAFOR - BAAGN)', data=Teams4).fit()
WinBA2_lm.summary()
```
