```python
# First we load the packages
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

# I. Building the Dataset: Player Stats

```python
# Read in salary data retrieved from Lahman's Database
Salary = pd.read_csv("D:\\GitHub\\Coursera\\SportsPerformanceAnalytics\\MoneyBallAndBeyond\\data\\Salaries.csv")

# Read in batting
Batting = pd.read_csv("D:\\GitHub\\Coursera\\SportsPerformanceAnalytics\\MoneyBallAndBeyond\\data\\Batting.csv")
```

```python
#Your Code Here
Salary = Salary[Salary.salary > 0] 
Salary
```

```python
Salary['lnSal'] = np.log(Salary['salary'])
Salary = Salary.rename(columns = {'yearID':'SalYear'})
Salary
```

```python
Master = Salary
Master
```

```python
Batting = Batting.groupby(['playerID','yearID']).sum()
Batting.reset_index(inplace=True)
Batting
```

```python
Batting = Batting[(Batting.yearID >= 1998) & (Batting.yearID <= 2006) & (Batting.AB >= 130)].copy()
Batting
```

```python
Batting['PA'] = Batting['AB'] + Batting['BB'] + Batting['HBP'] + Batting['SH'] + Batting['SF']

Batting['OBP'] = (Batting['H'] + Batting['BB'] + Batting['HBP'])/(Batting['AB'] + Batting['BB']
                                                                  + Batting['HBP'] + Batting['SF'])

Batting['SLG'] = ((Batting['H'] - Batting['Doubles'] - Batting['Triples'] - Batting['HR']) 
                  + 2*Batting['Doubles'] + 3*Batting['Triples'] + 4*Batting['HR'])/Batting['AB']

display(Batting)
```

```python
Batting['SalYear'] = Batting['yearID'] + 1 
display(Batting)
```

```python
Master = pd.merge(Batting, Master, on=['SalYear', 'playerID'])
display(Master)
```

```python
Sal_1999 = Master[(Master.yearID == 1999)].copy()
Sal_1999
```

```python
Sal_2006 = Master[(Master.yearID == 2006)].copy()
Sal_2006
```

```python
avg_1999 = Sal_1999['salary'].sum()/Sal_1999['salary'].count()
avg_1999
```

```python
avg_2006 = Sal_2006['salary'].sum()/Sal_2006['salary'].count()
avg_2006
```

# II. Building the Dataset: Player Info

```python
# Read in People data retrieved from Lahman's Database
People = pd.read_csv("D:\\GitHub\\Coursera\\SportsPerformanceAnalytics\\MoneyBallAndBeyond\\data\\People.csv")

# Read in player appearance data retrieved from Lahman's Database
Appearances = pd.read_csv("D:\\GitHub\\Coursera\\SportsPerformanceAnalytics\\MoneyBallAndBeyond\\data\\Appearances.csv")
```

```python
#Your Code Here
People
```

```python
Debut = People[['playerID','debut']].copy()
```

```python
Debut['debutyr'] = Debut['debut'].astype(str).str[0:4]
Debut
```

```python
Debut = Debut[['playerID','debutyr']]
Master = pd.merge(Master, Debut, on=['playerID'], how = 'left')
display(Master)
```

```python
Master['Exp'] = Master['yearID'] - Master['debutyr'].astype(int)
display(Master)
```

```python
Master['Arb'] = np.where((Master['Exp'] <= 6) & (Master['Exp'] >= 3),1,0)
Master['Free'] = np.where(Master['Exp'] > 6, 1, 0)
display(Master)
```

```python
Appearances
```

```python
Appearances = Appearances.groupby(['playerID','yearID'])['G_c','G_1b','G_2b','G_3b','G_ss','G_of','G_dh'].sum()                                       
Appearances.reset_index(inplace=True)
display(Appearances)
```

```python
Appearances['Max_G'] = Appearances[["G_c","G_1b","G_2b","G_3b","G_ss","G_of","G_dh"]].max(axis=1)
display(Appearances)
```

```python
def Position(df):
    if (df['Max_G'] == df['G_c']): return "C"
    elif (df['Max_G'] == df['G_1b']): return "1B"
    elif (df['Max_G'] == df['G_2b']): return "2B"
    elif (df['Max_G'] == df['G_3b']): return "3B"
    elif (df['Max_G'] == df['G_ss']): return "SS"
    elif (df['Max_G'] == df['G_of']): return "OF"
    elif (df['Max_G'] == df['G_dh']): return "DH"
```

```python
Appearances['POS'] = Appearances.apply(Position, axis = 1)    

display(Appearances)
```

```python
Appearances = Appearances[Appearances['Max_G'] > 0] 
Appearances = Appearances[['playerID','yearID','POS']]
display(Appearances)
```

```python
Appearances['Catch'] = np.where(Appearances['POS'] == "C", 1, 0)
Appearances['Infld'] = np.where((Appearances['POS'] == "2B") | (Appearances['POS'] == "3B") | 
                                (Appearances['POS'] == "SS"), 1, 0)
Appearances['B2'] = np.where((Appearances['POS'] == "2B"),1,0)
Appearances['B3'] = np.where((Appearances['POS'] == "3B"),1,0)
Appearances['SS'] = np.where((Appearances['POS'] == "SS"),1,0)

display(Appearances)
```

```python
Master = pd.merge(Master, Appearances, on=['playerID','yearID'], how = 'left')
display(Master)
```

```python
#### Quiz Work ####
pd.set_option('display.max_rows', None)
Money = Master.groupby(['yearID','POS']).sum()/Master.groupby(['POS']).count()
Money.to_csv('Salary.csv')
```

# III. Running Regressions

```python
MB_Data = Master[(Master.SalYear >= 1999) & (Master.SalYear <= 2003)]
MB_Data['BA'] = MB_Data['H']/MB_Data['AB']
MB_Data
```

```python
#Your Code Here
import statsmodels.formula.api as smf
Val_All_lm = smf.ols(formula = 'lnSal ~ OBP + SLG + BA + PA + Arb + Free + Catch + B2 + B3 + SS', data=MB_Data).fit()
Val_All_lm.summary()
```

```python
After = Master[(Master.SalYear >= 2004) & (Master.SalYear <= 2006)]
After['BA'] = After['H']/After['AB']
After
```

```python
Val_Af_lm = smf.ols(formula = 'lnSal ~ OBP + SLG + BA + PA + Arb + Free + Catch + B2 + B3 + SS', data=After).fit()
Val_Af_lm.summary()
```

```python
Master['BA'] = Master['H']/Master['AB']
```

```python
MB_Data_1999 = Master[(Master.SalYear == 1999)]
Val_1999_lm = smf.ols(formula = 'lnSal ~ OBP + SLG + BA + PA + Arb + Free + Catch + B2 + B3 + SS', data=MB_Data_1999).fit()
MB_Data_2000 = Master[(Master.SalYear == 2000)]
Val_2000_lm = smf.ols(formula = 'lnSal ~ OBP + SLG + BA + PA + Arb + Free + Catch + B2 + B3 + SS', data=MB_Data_2000).fit()
MB_Data_2001 = Master[(Master.SalYear == 2001)]
Val_2001_lm = smf.ols(formula = 'lnSal ~ OBP + SLG + BA + PA + Arb + Free + Catch + B2 + B3 + SS', data=MB_Data_2001).fit()
MB_Data_2002 = Master[(Master.SalYear == 2002)]
Val_2002_lm = smf.ols(formula = 'lnSal ~ OBP + SLG + BA + PA + Arb + Free + Catch + B2 + B3 + SS', data=MB_Data_2002).fit()
MB_Data_2003 = Master[(Master.SalYear == 2003)]
Val_2003_lm = smf.ols(formula = 'lnSal ~ OBP + SLG + BA + PA + Arb + Free + Catch + B2 + B3 + SS', data=MB_Data_2003).fit()
```

```python
from statsmodels.iolib.summary2 import summary_col
Header = ['1999','2000','2001','2002','2003']
Table_Pre = summary_col([Val_1999_lm,Val_2000_lm,Val_2001_lm,Val_2002_lm,Val_2003_lm,],\
                      regressor_order=['OBP','SLG','BA','PA','Arb','Free','Catch','B2','B3','SS','Intercept'],stars=True, \
                      float_format="'%.3f'",model_names = Header)
print(Table_Pre)
```

```python
After_2004 = Master[(Master.SalYear == 2004)]
After_2004_lm = smf.ols(formula = 'lnSal ~ OBP + SLG + BA + PA + Arb + Free + Catch + B2 + B3 + SS', data=After_2004).fit()
After_2005 = Master[(Master.SalYear == 2005)]
After_2005_lm = smf.ols(formula = 'lnSal ~ OBP + SLG + BA + PA + Arb + Free + Catch + B2 + B3 + SS', data=After_2005).fit()
After_2006 = Master[(Master.SalYear == 2006)]
After_2006_lm = smf.ols(formula = 'lnSal ~ OBP + SLG + BA + PA + Arb + Free + Catch + B2 + B3 + SS', data=After_2006).fit()
```

```python

Header = ['2004','2005','2006']
Table_Post = summary_col([After_2004_lm,After_2005_lm,After_2006_lm,],\
                      regressor_order=['OBP','SLG','BA','PA','Arb','Free','Catch','B2','B3','SS','Intercept'],stars=True, \
                      float_format="'%.3f'",model_names = Header)
print(Table_Post)
```
